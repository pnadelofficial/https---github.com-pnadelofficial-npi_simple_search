```python
import pandas as pd
from whoosh.fields import Schema, TEXT, ID
from whoosh import index
import os, os.path
from whoosh import index
from whoosh import qparser
from whoosh.qparser import QueryParser
```

```python
schema = Schema(org= ID(stored=True), path=ID(stored=True), content=TEXT(stored=True))
```

```python
if not os.path.exists("transcripts_index_dir"):
    os.mkdir("transcripts_index_dir")
```

```python
at = pd.read_csv('all_transcripts.csv').dropna().rename(columns={'index':'org_index'})
```

```python
at
```

```python
ix = index.create_in("transcripts_index_dir", schema)
writer = ix.writer()

for i in range(len(at)):
    writer.add_document(org=str(at.org_index.iloc[i]), content=str(at.q_a.iloc[i]), path=str(i))

writer.commit()
```

```python
if not os.path.exists("statements_index_dir"):
    os.mkdir("statements_index_dir")
```
